### Question 1:
Given the following two documents:

- Document 1: "The quick brown fox jumps over the lazy dog."

- Document 2: "The fox."

**What is the TF of the word 'fox' in Document 1?**

## Solution:
TF (Term Frequency) is calculated as the number of times a term appears in a document divided by the total number of terms in the document.

For Document 1:

The term 'fox' appears 1 time.
Total number of terms in Document 1: 9
